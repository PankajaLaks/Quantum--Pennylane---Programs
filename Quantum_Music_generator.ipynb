```python
!pip install pennylane
import pennylane as qml
import numpy as np
import random

# 2 qubits can produce 4 outputs: 00, 01, 10, 11
dev = qml.device("default.qubit", wires=2, shots=1)

@qml.qnode(dev)
def quantum_note():
    qml.Hadamard(wires=0)
    qml.Hadamard(wires=1)
    return qml.sample(wires=[0, 1])

# Map quantum outputs to music notes
note_map = {
    "00": "C",
    "01": "D",
    "10": "E",
    "11": "G"
}

melody = []

# Generate 16 quantum notes
for i in range(16):
    result = quantum_note()
    # Access the single shot's measurements (the inner array) before iterating
    bit_string = "".join(str(int(bit)) for bit in result[0])
    note = note_map[bit_string]
    melody.append(note)

print("馃幍 Quantum Generated Melody:")
print(" - ".join(melody))
```
