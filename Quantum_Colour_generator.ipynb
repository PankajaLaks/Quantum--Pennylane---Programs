```python
# Install:
!pip install pennylane matplotlib

import pennylane as qml
import matplotlib.pyplot as plt

# 8 qubits for R, 8 for G, 8 for B = 24 qubits
dev = qml.device("default.qubit", wires=24, shots=1)

@qml.qnode(dev)
def quantum_rgb_bits():
    for i in range(24):
        qml.Hadamard(wires=i)
    return qml.sample(wires=range(24))

def bits_to_int(bits):
    bit_string = "".join(str(int(b)) for b in bits)
    return int(bit_string, 2)

# Generate quantum bits
# Access the single shot's measurements (the inner array)
bits = quantum_rgb_bits()[0]

# Split into RGB
r_bits = bits[0:8]
g_bits = bits[8:16]
b_bits = bits[16:24]

r = bits_to_int(r_bits)
g = bits_to_int(g_bits)
b = bits_to_int(b_bits)

print("Quantum RGB Color:", (r, g, b))

# Display color
color = [[(r / 255, g / 255, b / 255)]]

plt.imshow(color)
plt.axis("off")
plt.title(f"Quantum Color RGB = ({r}, {g}, {b})")
plt.show()
```
